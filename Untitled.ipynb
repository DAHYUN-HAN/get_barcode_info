```python
import pyzbar.pyzbar as pyzbar
import cv2
import matplotlib.pyplot as plt
import os
```

```python
import pandas as pd
import numpy as np
```

```python
from tqdm import tqdm
```

```python
file_path = 'book_barcod_augmentation_transform_barcode_crop/'
file_names = os.listdir(file_path)
result_list = []
for name in file_names:
#     print(name)
#     with open('a.txt', 'a') as f:
#         f.writelines(name+'\n')
    src = os.path.join(file_path, name)
    img = cv2.imread(src)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    decoded = pyzbar.decode(gray)
    for d in decoded:

#         print(d.data.decode('utf-8'))
        result_list.append([name, d.data.decode('utf-8')])
        cv2.rectangle(img, (d.rect[0], d.rect[1]), (d.rect[0] + d.rect[2], d.rect[1] + d.rect[3]), (0, 0, 255), 2)
#     cv2.imwrite('images/'+name, img)
```

```python
file_path = 'book_barcod_augmentation_transform_barcode_crop/'
file_names = os.listdir(file_path)
result_list = []
for name in file_names:
    src = os.path.join(file_path, name)
    img = cv2.imread(src)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    decoded = pyzbar.decode(gray)

    right = 0
    if(len(decoded) == 0):
        result_list.append([name, ' '])
        
    elif(len(decoded) == 1):
        result_list.append([name, decoded[0].data.decode('utf-8')])
        cv2.rectangle(img, (decoded[0].rect[0], decoded[0].rect[1]), (decoded[0].rect[0] + decoded[0].rect[2],decoded[0].rect[1] + decoded[0].rect[3]), (0, 0, 255), 2)
        
    elif(len(decoded) != 1):
        for i in range(len(decoded)):
            if(len(decoded[i].data) > len(decoded[right].data)):
                right = i
        result_list.append([name, decoded[right].data.decode('utf-8')])
        cv2.rectangle(img, (decoded[right].rect[right], decoded[right].rect[1]), (decoded[right].rect[right] + decoded[right].rect[2],decoded[right].rect[1] + decoded[right].rect[3]), (0, 0, 255), 2)
    cv2.imwrite('images2/'+name, img)
```

```python
df = pd.DataFrame(result_list)
df.to_excel('result2.xlsx')
```
